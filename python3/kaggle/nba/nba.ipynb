## NBA最佳球员得分分析（kaggle）
>NBA球员的防守表现是否比进攻表现更适合更适合用来预测球员的RPM?


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
play_df = pd.read_csv('./nba_2017_nba_players_with_salary.csv')
```

```python
play_df.head()
```

```python
# sns.set() # 增加底纹
plt.figure(figsize=(15,7))
sns.regplot(x='MP',y='WINS_RPM', data = play_df)
plt.show()
```

```python
# sns.set() # 增加底纹

ax = sns.lmplot(x='MP'
                ,y = 'WINS_RPM'
                ,data=play_df
                ,hue='POSITION'
                ,fit_reg=False
                ,height=6
                ,aspect=2
                ,legend=False
                ,scatter_kws={"s":200}
               )
ax.set( xlabel = 'Minutes played'
      ,ylabel = 'RPM(Real PLus Minus)'
      ,title = "Minutes played Vs RPM by Position:2016-2017 Season")

plt.legend(loc='upper left'
          ,title = 'Position')
plt.show()
```
